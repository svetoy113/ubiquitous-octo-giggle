```python
def load_and_prepare_data(sentences):
    # Load the CSV
    data = pd.DataFrame(sentences, columns=['Sentence'])
    
    # Manually add some 'Other' data (this is just an example)
    additional_data = [
        {"Quote": "It might rain tomorrow.", "Author": "Unknown"},
        {"Quote": "I am not sure about the future.", "Author": "Unknown"},
        # Add more as needed...
    ]
    additional_df = pd.DataFrame(additional_data)
    additional_df['Label'] = 'Other'
    
    # Add labels to our positive quotes
    data['Label'] = 'Positive'

    # Combine datasets
    combined_data = pd.concat([data, additional_df], ignore_index=True)
    # Replace NaN values with an empty string
    combined_data.fillna("", inplace=True)
    
    # return combined_data
    return combined_data

def train_model(data):
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(data['Quote'], data['Label'], test_size=0.2, random_state=42)

    # Vectorize text
    vectorizer = CountVectorizer()
    X_train_counts = vectorizer.fit_transform(X_train)
    X_test_counts = vectorizer.transform(X_test)

    # Train a Naive Bayes classifier
    clf = MultinomialNB()
    clf.fit(X_train_counts, y_train)

    # Predict and evaluate
    y_pred = clf.predict(X_test_counts)
    print(classification_report(y_test, y_pred))

def main_train():
    
    data = load_and_prepare_data(sentences)
    train_model(data)

while True:     
    # Generate 10,000 sentences
    num_sentences = 10000
    sentence_length = 7  # You can adjust the length of each sentence
    sentences = [generate_random_sentence(sentence_length) for _ in range(num_sentences)]
    main_train()
```
